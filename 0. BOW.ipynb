# 数据概览

```python
data = pd.read_excel('data/广东(数据清洗4).xlsx', index_col=0)
data.drop(columns=['景点','昵称','等级','时间'], inplace=True)
data.columns = ['Body', 'Rank']
```

看下每个星级的评论内容：

```python
for rank in range(0, 6):
    print('\nRank: ',rank)
    print(data[data['Rank']==rank]['Body'].head(5))
```

可以看见，0～3 星基本是负面评价，而 4～5 星基本是正面评价，因此我们将按此进行二分类：

```python
data['Rank'] = data['Rank'].map({0:0, 1:0, 2:0, 3:0, 4:1, 5:1})
data.head()
data.groupby('Rank').count().plot.bar()
data.groupby('Rank').count()
```

可以看见，结果划分后，两类评价的数量基本一致。

# 结巴分词

```python
import jieba
```

```python
data_split = data.copy()
data_split.columns = ['Body', 'Sentiment']
data_split['Body'] = data_split['Body'].map(lambda x:jieba.lcut(x))
data_split.head()
```

# 停用词

```python
# 读取停用词表，剔除停用词
s = ''
with open('data/stopwords.txt', 'r', encoding='utf8') as r:
    for i in r.readlines():
        s += i.strip()
data_split['Body'] = data_split['Body'].map(lambda x:[i for i in x if(i not in s) and (len(i) > 1)])
data_split.head()
```

```python
s
```

# 向量化

```python
from sklearn.feature_extraction.text import CountVectorizer
# 将剔除停用词后的Body特征转换为矩阵向量
list1 = data_split['Body'].map(lambda line:' '.join(line))
cv = CountVectorizer()
w = cv.fit_transform(list1)
```

# 降维前

```python
from sklearn import model_selection, metrics
from sklearn.naive_bayes import BernoulliNB
```

```python
Y = data_split['Sentiment']
train_x, test_x, train_y, test_y = \
model_selection.train_test_split(w, Y, test_size=0.2, random_state=0)

# 建立朴素贝叶斯分类模型
model = BernoulliNB()
model.fit(train_x, train_y)
# 打印混淆矩阵和分类报告
print('classification_report on test set:\n', 
      metrics.classification_report(test_y, model.predict(test_x)))

print('classification_report on train set:\n', 
      metrics.classification_report(train_y, model.predict(train_x)))
```

# LSA

```python
from sklearn.decomposition import TruncatedSVD
# 降维处理得到特征X
svd = TruncatedSVD(200)
X = svd.fit_transform(w)
print(sum(svd.explained_variance_ratio_))
# 切片sentiment特征作为目标标签Y
Y = data_split['Sentiment']
train_x, test_x, train_y, test_y = \
model_selection.train_test_split(X, Y, test_size=0.2, random_state=0)
```
